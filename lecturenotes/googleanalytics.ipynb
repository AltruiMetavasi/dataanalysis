```python
import pandas as pd
print(pd.__version__)
```

```python
ana = pd.read_csv("data_input/analytics.csv", comment="#").head(11)
ana.head()
```

```python
ana.Users = ana.Users.str.replace(',', '')
ana.Users = ana.Users.astype('int')

ana['New Users'] = ana['New Users'].str.replace(',', '')
ana['New Users'] = ana['New Users'].astype('int')
```

```python
ana['HiVol'] = ana['New Users']/ana['Users'] > 0.95
ana.tail()
```

```python
ana = ana.fillna("notprovided")
cond1 = ana.Language.str.startswith('en')
cond2 = ana.Language.str.startswith('id')
ana.loc[cond1, 'Language'] = 'en'
ana.loc[cond2, 'Language'] = 'id'
```

```python
ana.head(3)
```

```python
pd.crosstab(index=ana.Language, 
            columns=ana.HiVol,
            values=ana['Pages / Session'],
            aggfunc='sum',
            margins=True
           ).fillna(0).round(2)
```

```python
pd.pivot_table(data=ana, 
               index='Language', 
               columns='HiVol', 
               values='Pages / Session')
```
